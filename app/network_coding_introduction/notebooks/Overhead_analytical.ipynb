# Overhead due to linear dependencies

Lets import a module that allows us to plot graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

In the book chapter, we stated that the overhead due to linear dependencies of network coding is equal to
$$O_T =  \left(\sum_{r'=1}^{g} \frac{1}{1-q^{-r'}} \right) - g \:, $$
which is a function of the size of the Galois field $q$, and the generation size $g$.

Lets define a function to compute the overheads for $g=1$ to $g = 16$ for a $q$ provided by the user.

```python
def ComputeOverheads(q):
    overheads = []
    for g in range(1, 17):
        Ot = 0
        for r in range(1,g+1):
            Ot += 1.0 / (1.0 - q**(-r))
        overheads.append(Ot - g)
    return overheads
```

## Overhead of the binary field

Lets define $q = 2$ (the binary field)

```python
q = 2
overheads = ComputeOverheads(q)
```

```python
# plt.rc('text', usetex=True)
plt.rc('font', family='serif', size = 19)

plt.bar(range(1,17),overheads)
plt.xlabel("Generation size $g$",size=24)
plt.ylabel("Overhead",size=24)
plt.xticks(range(0,17,2))
plt.yticks(np.arange(0,2,0.4))
plt.grid()
plt.tight_layout()
# plt.savefig('overhead_analytical_q_2.png',dpi=600)
```

## Overhead of the binary8 field

The overhead for a $GF(2^8)$ is:

```python
q = 2**8
overheads = ComputeOverheads(q)
```

```python
# plt.rc('text', usetex=True)
plt.rc('font', family='serif', size = 19)

plt.bar(range(1,17),overheads)
plt.xlabel("Generation size $g$",size=24)
plt.ylabel("Overhead",size=24)
plt.xticks(range(0,17,2))
plt.grid()
plt.tight_layout()
# plt.savefig('overhead_analytical_q_256.png',dpi=600)
```
